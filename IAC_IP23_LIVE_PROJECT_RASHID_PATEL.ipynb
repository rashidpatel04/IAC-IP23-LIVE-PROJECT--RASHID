# Q5.What is the average family income of the student?


```python
# Function to clean and convert income strings to numerical values
import re
def clean_and_convert_income(income_str):
    # Use regular expression to extract numerical part from string
    match = re.search(r'\d+', str(income_str))  # Convert to string in case it's not
    if match:
        return int(match.group())
    else:
        return None  # Return None if no numerical value found (handle error cases)

# Apply the cleaning function to 'family_income' column
data['Family Income'] = data['Family Income'].apply(clean_and_convert_income)

# Calculate the average family income
average_income = data['Family Income'].mean()

print(f"The average family income of the students is ₹{average_income:.2f} lakhs")
```

# Q9.Can we identify any relationship between family income and GPA?

```python

import seaborn as sns
import matplotlib.pyplot as plt
# Convert GPA to numeric
data['CGPA'] = pd.to_numeric(data['CGPA'], errors='coerce')

# Map family income categories to numerical values
income_mapping = {
    '0-2 Lakh': 1,
    '2-5 Lakh': 2,
    '5-7 Lakh': 3,
    '7 Lakh+': 4
}
data['Family Income'] = data['Family Income'].map(income_mapping)

# Drop rows with missing values in relevant columns
data_clean = data.dropna(subset=['CGPA', 'Family Income'])

# Calculate the correlation coefficient
correlation = data_clean['Family Income'].corr(data_clean['CGPA'])
print(f"Correlation between Family Income and GPA: {correlation}")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Family Income', y='CGPA', data=data_clean)
plt.title('Scatter Plot of Family Income vs GPA')
plt.xlabel('Family Income (Numerical)')
plt.ylabel('GPA')
plt.show()
```

# Q10.How does the expected salary vary based on factors like 'GPA', 'Family income', 'Experience with python (Months)'?


```python
#not done yet
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Convert columns to appropriate data types
data['Expected salary (Lac)'] = pd.to_numeric(data['Expected salary (Lac)'], errors='coerce')
data['CGPA'] = pd.to_numeric(data['CGPA'], errors='coerce')
data['Experience with python (Months)'] = pd.to_numeric(data['Experience with python (Months)'], errors='coerce')

# Drop rows with missing values in relevant columns
data_clean = data.dropna(subset=['Expected salary (Lac)', 'CGPA', 'Family Income', 'Experience with python (Months)'])

# Mapping family income to numerical values for analysis
income_mapping = {
    '0-2 Lakh': 1,
    '2-5 Lakh': 2,
    '5-7 Lakh': 3,
    '7 Lakh+': 4
}
data_clean['Family Income'] = data_clean['Family Income'].map(income_mapping)

# Correlation matrix
correlation_matrix = data_clean[['Expected salary (Lac)', 'CGPA', 'Family Income', 'Experience with python (Months)']].corr()
print("Correlation Matrix:\n", correlation_matrix)

# Scatter plots
sns.pairplot(data_clean[['Expected salary (Lac)', 'CGPA', 'Family Income', 'Experience with python (Months)']])
plt.show()

# Regression analysis
X = data_clean[['CGPA', 'Family Income', 'Experience with python (Months)']]
X = sm.add_constant(X)  # Adding a constant term for the intercept
y = data_clean['Expected salary (Lac)']

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

print(model.summary())
```
